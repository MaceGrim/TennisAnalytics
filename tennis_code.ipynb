```python
rankings = pd.read_excel('tennis_rankings.xls')
rankings['file_name'] = rankings['Last Name'] + '_' + rankings['Rank'].astype(str)
rankings['xls'] = '('+ rankings['Rank'].astype(str) + ') ' + rankings['Last Name'] + '.xls'
```

```python
df = pd.DataFrame()
files_xls = rankings['xls'].tolist()
```

```python
for f in files_xls:
    data = pd.read_excel(f)
    df = df.append(data)
```

```python
df[df['Player 1'].str.contains('Nadal') & df['Player 2'].str.contains('Murray')]
```

```python
df_clean = df.drop_duplicates(subset = ['Player 1', 'Player 2', 'Tournament', 'Result'])
```

```python
df_clean[df_clean['Player 1'].str.contains('Nadal') & df_clean['Player 2'].str.contains('Murray')]
```

```python
player_matrix = pd.DataFrame(index = rankings['Last Name'].tolist(),columns = rankings['Last Name'].tolist())
player_rankings_list = rankings['Last Name'].tolist()
```

```python
player_matrix = player_matrix[player_matrix.columns[::-1]]
for m in (player_rankings_list):
    for n in (player_rankings_list):
        temp_chrono = pd.DataFrame(df_clean[(df_clean['Player 1'].str.contains(m) &  df_clean['Player 2'].str.contains(n)) | (df_clean['Player 1'].str.contains(n) &  df_clean['Player 2'].str.contains(m))]['Player 1'].str.contains(m).astype(int))
        player_matrix[m][n] = temp_chrono['Player 1'].tolist()
```

```python
player_matrix
```

```python
player_matrix_chronological = pd.DataFrame(index = rankings['Last Name'].tolist(),columns = rankings['Last Name'].tolist())
for m in (player_rankings_list):
    for n in (player_rankings_list):
        player_matrix_chronological[m][n] = player_matrix[n][m]
player_matrix_chronological = player_matrix_chronological[player_matrix_chronological.columns[::-1]]
player_matrix_chronological
```

```python
player_matrix_chronological.to_excel('player_matrix_chronological.xlsx')
```

```python
df_clean.to_excel('player_data_merged.xlsx')
```
